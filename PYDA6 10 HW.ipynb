### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
# "ratings.dat"
# UserID::MovieID::Rating::Timestamp
# "users.dat"
# UserID::Gender::Age::Occupation::Zip-code
# "movies.dat"
# MovieID::Title::Genres
```

```python
#готовим таблицы
ratings = pd.read_csv(filepath_or_buffer='ml-1m/ratings.dat', delimiter=r'::', header=None, names=['UserID','MovieID','Rating','Timestamp'], usecols=[1,2], dtype=int, engine='python')
bestratings = ratings[ratings['Rating']==5]
ratings = None
movies = pd.read_csv(filepath_or_buffer='ml-1m/movies.dat', delimiter=r'::', header=None, names=['MovieID','Title','Genres'], usecols=[0,1], engine='python')
```

```python
bestmovie = bestratings['MovieID'].value_counts().keys()[0] #value_counts() посчитал количество строк в разрезе и отсортировал по убыванию, keys()[0] дает ИД фильма
movies[movies['MovieID']==bestmovie].iloc[0]['Title']
```

### Задание 2

```python
power = pd.read_csv('power.csv')
#country,year,quantity,category
countries = ['Latvia', 'Lithuania', 'Estonia']
categories = [4, 12, 21]
periods = list(range(2005,2011))
```
