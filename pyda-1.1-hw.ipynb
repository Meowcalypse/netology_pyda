**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

```python
months = ('январь','февраль','март','апрель','май','июнь','июль','август','сентябрь','октябрь','ноябрь','декабрь') #месяцы
days = (20,19,21,20,21,21,23,23,23,23,22,22) #даты смены знаков по месяцам
signs = ('козерог','водолей','рыбы','овен','телец','близнецы','рак','лев','дева','весы','скорпион','стрелец','козерог') #знаки на начало месяца
u_month = input('Введите месяц:')
u_day = int(input('Введите число:'))

try:
    t_monthno = months.index(u_month.lower()) #номер месяца
    SignIndex = t_monthno #номер знака
    if days[t_monthno]<=u_day:
        SignIndex = (t_monthno +1)%12 #смещение номера знака в зависимости от даты; остаток от деления покрывает последние числа декабря
    print(signs[SignIndex])
except Exception as err:
    print('Ошибка ввода данных')
    print(format(err))
```

**Задание 4**. Нужно разработать приложение для финансового планирования.
Приложение учитывает сколько уходит на ипотеку, "на жизнь" и сколько нужно отложить на пенсию.
Пользователь вводит:
- заработанную плату в месяц.
- сколько процентов от ЗП уходит на ипотеку.
- сколько процентов от ЗП уходит "на жизнь".
- сколько раз приходит премия в год.

Остальная часть заработанной платы откладывается на пенсию.

Также пользователю приходит премия в размере зарплаты, от которой половина уходит на отпуск, а вторая половина откладывается.

Программа должна учитывать сколько премий было в год.

Нужно вывести сколько денег тратит пользователь на ипотеку и сколько он накопит за год.

Пример:

Введите заработанную плату в месяц: 100000

Введите сколько процентов уходит на ипотеку: 30

Введите сколько процентов уходит на жизнь: 50

Введите количество премий за год: 2

Вывод:
На ипотеку было потрачено: 360000 рублей
Было накоплено: 340000 рублей

```python
u_salary = int(input('Введите месячную ЗП:'))
u_credit_rate = int(input('Введите сколько процентов уходит на ипотеку:'))
u_life_rate = int(input('Введите сколько процентов уходит на жизнь:'))
u_bonus_count = int(input('Введите количество премий за год:'))

credit_spent = u_salary*12*u_credit_rate/100 #ЗП * 12 * процент трат на ипотеку
accumulated = u_salary*(u_bonus_count + 12*(100-u_life_rate)/100) - credit_spent # ЗП * количество бонусов + ЗП * 12 месяцев * остаток от трат на жизнь - затраты на ипотеку
print('На ипотеку было потрачено: ' +f'{credit_spent:.0f}'+ ' рублей Было накоплено: '+f'{accumulated:.0f}'+' рублей')
```
