**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

```python
months = ('январь','февраль','март','апрель','май','июнь','июль','август','сентябрь','октябрь','ноябрь','декабрь')
days = (20,19,21,20,21,21,23,23,23,23,22,22)
signs = ('козерог','водолей','рыбы','овен','телец','близнецы','рак','лев','дева','весы','скорпион','стрелец','козерог')
u_month = input('Введите месяц:')
u_day = int(input('Введите число:'))

try:
    t_monthno = months.index(u_month.lower())
    SignIndex = t_monthno
    if days[t_monthno]<=u_day:
        SignIndex = t_monthno +1
    print(signs[SignIndex])
except Exception as err:
    print('Ошибка ввода данных')
    print(format(err))
```
